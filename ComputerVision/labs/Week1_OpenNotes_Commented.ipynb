```python
# Core libraries used across this notebook. Run this cell first.
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ✅ Change this to your image file name
# Store the result in 'IN_PATH' so we can use it in later lines.
IN_PATH = "input.jpg"  # example: "cat.jpg", "face_noise.png", etc.
```

```python
# ----------------------------
# 1) Reading images
```

```python
# ----------------------------

# Define a reusable function named 'read_color' so we can run this block later when needed.
def read_color(path):
    """
    Read a color image from disk (OpenCV loads it as BGR).
    Returns: BGR image as a NumPy array with shape (H, W, 3)
    """
    # Load an image file from disk into memory.
    img = cv2.imread(path, cv2.IMREAD_COLOR)
    # Check this condition (img is None); run the indented lines only when it is true.
    if img is None:
        # Stop here and show an error message so problems are caught early.
        raise ValueError(f"Could not read image: {path}")
    # Send this value back as the function's result.
    return img


# Define a reusable function named 'read_gray' so we can run this block later when needed.
def read_gray(path):
    """
    Read an image as grayscale (single channel).
    Returns: gray image as a NumPy array with shape (H, W)
    """
    # Load an image file from disk into memory.
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    # Check this condition (img is None); run the indented lines only when it is true.
    if img is None:
        # Stop here and show an error message so problems are caught early.
        raise ValueError(f"Could not read image: {path}")
    # Send this value back as the function's result.
    return img
```

```python
# ----------------------------
# 2) Showing images (Matplotlib)
```

```python
# ----------------------------

# Define a reusable function named 'bgr_to_rgb' so we can run this block later when needed.
def bgr_to_rgb(img_bgr):
    """Convert BGR -> RGB for correct display in matplotlib."""
    # Send this value back as the function's result.
    return cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)


# Define a reusable function named 'show_color' so we can run this block later when needed.
def show_color(img_bgr, title="Color Image"):
    """Show a BGR image correctly using matplotlib."""
    # Start a new blank chart window before drawing anything.
    plt.figure()
    # Draw the image so it appears in the chart window.
    plt.imshow(bgr_to_rgb(img_bgr))
    # Set the text title shown above the image or chart.
    plt.title(title)
    # Control whether axis marks are visible (here they are hidden).
    plt.axis("off")
    # Display the chart or image on screen now.
    plt.show()


# Define a reusable function named 'show_gray' so we can run this block later when needed.
def show_gray(img_gray, title="Gray Image"):
    """Show a grayscale image using matplotlib."""
    # Start a new blank chart window before drawing anything.
    plt.figure()
    # Draw the image so it appears in the chart window.
    plt.imshow(img_gray, cmap="gray")
    # Set the text title shown above the image or chart.
    plt.title(title)
    # Control whether axis marks are visible (here they are hidden).
    plt.axis("off")
    # Display the chart or image on screen now.
    plt.show()
```

```python
# ----------------------------
# 3) Saving images
```

```python
# ----------------------------

# Define a reusable function named 'save_image' so we can run this block later when needed.
def save_image(path, img):
    """Save an image to disk. Returns True/False in OpenCV."""
    # Write the image from memory out to a file on disk.
    ok = cv2.imwrite(path, img)
    # Check this condition (not ok); run the indented lines only when it is true.
    if not ok:
        # Stop here and show an error message so problems are caught early.
        raise ValueError(f"Could not save image: {path}")
    # Send this value back as the function's result.
    return ok
```

```python
# ----------------------------
# 4) Grayscale conversion (manual)
```

```python
# ----------------------------

# Define a reusable function named 'manual_grayscale' so we can run this block later when needed.
def manual_grayscale(img_bgr):
    """
    Convert BGR -> grayscale manually.

    Common formula (luminance):
      gray = 0.299*R + 0.587*G + 0.114*B

    Because OpenCV is BGR:
      B = img[...,0], G = img[...,1], R = img[...,2]
    """
    # Store the result in 'h, w, _' so we can use it in later lines.
    h, w, _ = img_bgr.shape
    # Store the result in 'gray' so we can use it in later lines.
    gray = np.zeros((h, w), dtype=np.uint8)

    # Nested loops are clear for learning, but vectorized NumPy is faster for large images.
    # Repeat the indented steps for each item in this sequence.
    for i in range(h):
        # Repeat the indented steps for each item in this sequence.
        for j in range(w):
            # Store the result in 'b' so we can use it in later lines.
            b = img_bgr[i, j, 0]
            # Store the result in 'g' so we can use it in later lines.
            g = img_bgr[i, j, 1]
            # Store the result in 'r' so we can use it in later lines.
            r = img_bgr[i, j, 2]
            # Store the result in 'val' so we can use it in later lines.
            val = 0.299 * r + 0.587 * g + 0.114 * b
            # Run this step as written to continue the image-processing workflow.
            gray[i, j] = int(val)

    # Send this value back as the function's result.
    return gray
```

```python
# ----------------------------
# 5) Basic geometric operations
```

```python
# ----------------------------

# Define a reusable function named 'crop' so we can run this block later when needed.
def crop(img, y1, y2, x1, x2):
    """
    Crop using slicing.
    y = rows (height), x = cols (width)
    """
    # Send this value back as the function's result.
    return img[y1:y2, x1:x2]


# Define a reusable function named 'resize' so we can run this block later when needed.
def resize(img, new_w, new_h):
    """Resize to (new_w, new_h). OpenCV expects (width, height)."""
    # Send this value back as the function's result.
    return cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_AREA)


# Define a reusable function named 'flip_horizontal' so we can run this block later when needed.
def flip_horizontal(img):
    """Flip left <-> right."""
    # Send this value back as the function's result.
    return cv2.flip(img, 1)


# Define a reusable function named 'flip_vertical' so we can run this block later when needed.
def flip_vertical(img):
    """Flip up <-> down."""
    # Send this value back as the function's result.
    return cv2.flip(img, 0)


# Define a reusable function named 'rotate_90_clockwise' so we can run this block later when needed.
def rotate_90_clockwise(img):
    """Rotate 90 degrees clockwise (easy way)."""
    # Send this value back as the function's result.
    return cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
```

```python
# ----------------------------
# 6) Intensity operations (brightness/contrast)
```

```python
# ----------------------------

# Define a reusable function named 'clip_0_255' so we can run this block later when needed.
def clip_0_255(img_float):
    """
    Clip values into [0,255] and convert to uint8.
    """
    # Force all values to stay inside the safe range 0 to 255.
    img_float = np.clip(img_float, 0, 255)
    # Send this value back as the function's result.
    return img_float.astype(np.uint8)


# Define a reusable function named 'adjust_brightness' so we can run this block later when needed.
def adjust_brightness(gray, beta):
    """
    Brightness: add a constant.
    new = old + beta
    beta > 0 => brighter, beta < 0 => darker
    """
    # Send this value back as the function's result.
    return clip_0_255(gray.astype(np.float32) + beta)


# Define a reusable function named 'adjust_contrast' so we can run this block later when needed.
def adjust_contrast(gray, alpha):
    """
    Contrast: multiply by a constant.
    new = alpha * old
    alpha > 1 => more contrast, alpha < 1 => less contrast
    """
    # Send this value back as the function's result.
    return clip_0_255(gray.astype(np.float32) * alpha)


# Define a reusable function named 'adjust_contrast_brightness' so we can run this block later when needed.
def adjust_contrast_brightness(gray, alpha, beta):
    """
    Combined linear transform:
      new = alpha * old + beta
    """
    # Send this value back as the function's result.
    return clip_0_255(gray.astype(np.float32) * alpha + beta)
```

```python
# ----------------------------
# 7) Thresholding (binary image)
```

```python
# ----------------------------

# Define a reusable function named 'threshold_binary' so we can run this block later when needed.
def threshold_binary(gray, thresh):
    """
    Simple threshold:
      if pixel >= thresh => 255 (white)
      else              => 0   (black)
    """
    # Store the result in 'out' so we can use it in later lines.
    out = np.zeros_like(gray, dtype=np.uint8)
    # Run this step as written to continue the image-processing workflow.
    out[gray >= thresh] = 255
    # Send this value back as the function's result.
    return out
```

```python
# ----------------------------
# 8) Histogram (grayscale)
```

```python
# ----------------------------

# Define a reusable function named 'gray_histogram' so we can run this block later when needed.
def gray_histogram(gray):
    """
    Histogram counts how many pixels have intensity 0..255.
    Returns: hist (length 256)
    """
    # Flatten to 1D list of pixel values
    # Turn the image into one long list of pixel values.
    flat = gray.flatten()
    # Count how many times each pixel intensity appears.
    hist = np.bincount(flat, minlength=256)
    # Send this value back as the function's result.
    return hist


# Define a reusable function named 'plot_histogram' so we can run this block later when needed.
def plot_histogram(hist, title="Histogram"):
    # Start a new blank chart window before drawing anything.
    plt.figure()
    # Plot the data as a line graph.
    plt.plot(hist)
    # Set the text title shown above the image or chart.
    plt.title(title)
    # Label the horizontal axis.
    plt.xlabel("Intensity (0..255)")
    # Label the vertical axis.
    plt.ylabel("Pixel Count")
    # Display the chart or image on screen now.
    plt.show()
```

```python
# ----------------------------
# 9) Example usage (edit paths!)
```

```python
# ----------------------------
# Check this condition (__name__ == "__main__"); run the indented lines only when it is true.
# In notebooks, __name__ is typically "__main__", so this demo block runs when the cell executes.
if __name__ == "__main__":
    # Change this to whatever image you have
    # Store the result in 'in_path' so we can use it in later lines.
    in_path = "input.jpg"

    # Read
    # Store the result in 'color' so we can use it in later lines.
    color = read_color(in_path)
    # Convert image colors into a different format that we need.
    gray_cv = cv2.cvtColor(color, cv2.COLOR_BGR2GRAY)  # OpenCV built-in gray
    # Store the result in 'gray_manual' so we can use it in later lines.
    gray_manual = manual_grayscale(color)

    # Show
    # Run this step as written to continue the image-processing workflow.
    show_color(color, "Original (Color)")
    # Run this step as written to continue the image-processing workflow.
    show_gray(gray_cv, "Grayscale (OpenCV)")
    # Run this step as written to continue the image-processing workflow.
    show_gray(gray_manual, "Grayscale (Manual formula)")

    # Resize + crop
    # Store the result in 'small' so we can use it in later lines.
    small = resize(gray_cv, new_w=300, new_h=300)
    # Store the result in 'mid' so we can use it in later lines.
    mid = crop(small, 50, 250, 50, 250)
    # Run this step as written to continue the image-processing workflow.
    show_gray(mid, "Cropped")

    # Brightness + contrast
    # Store the result in 'brighter' so we can use it in later lines.
    brighter = adjust_brightness(gray_cv, beta=40)
    # Store the result in 'more_contrast' so we can use it in later lines.
    more_contrast = adjust_contrast(gray_cv, alpha=1.5)
    # Store the result in 'combo' so we can use it in later lines.
    combo = adjust_contrast_brightness(gray_cv, alpha=1.2, beta=20)

    # Run this step as written to continue the image-processing workflow.
    show_gray(brighter, "Brighter (beta=40)")
    # Run this step as written to continue the image-processing workflow.
    show_gray(more_contrast, "More Contrast (alpha=1.5)")
    # Run this step as written to continue the image-processing workflow.
    show_gray(combo, "Combo (alpha=1.2, beta=20)")

    # Threshold
    # Store the result in 'binary' so we can use it in later lines.
    binary = threshold_binary(gray_cv, thresh=120)
    # Run this step as written to continue the image-processing workflow.
    show_gray(binary, "Binary Threshold (thresh=120)")

    # Histogram
    # Store the result in 'hist' so we can use it in later lines.
    hist = gray_histogram(gray_cv)
    # Run this step as written to continue the image-processing workflow.
    plot_histogram(hist, "Histogram of Grayscale Image")

    # Save examples
    # Run this step as written to continue the image-processing workflow.
    save_image("week1_gray.png", gray_cv)
    # Run this step as written to continue the image-processing workflow.
    save_image("week1_binary.png", binary)
```
