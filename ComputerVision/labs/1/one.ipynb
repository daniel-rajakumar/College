# Image Processing Notebook (Easy Notes)

## Overview
This notebook shows how to load an image from the internet and study its colors.
We look at the color channels (Red, Green, Blue), and we also turn the image into grayscale in two ways.

## Topics Used
- OpenCV (`cv2`)
- NumPy arrays
- Matplotlib plotting
- Reading image data from a URL with `requests`
- BGR vs RGB color order
- Channel splitting and merging
- Custom grayscale formula vs OpenCV grayscale


## Step 1: Import libraries
We import tools for image loading, math, and drawing images.

```python
# Import OpenCV so we can process images
import cv2

# Import NumPy so we can work with number arrays
import numpy as np

# Import Matplotlib so we can display images
import matplotlib.pyplot as plt

# Import Requests so we can download the image from the internet
import requests
```

## Step 2: Download and decode the image
OpenCV needs image bytes converted into an array, then decoded into an image matrix.

```python
# Save the image link in a variable called url
url = "https://raymondmcguire.github.io/GPU-Based-Image-Processing-Tools/tools_webgl/FilterViewer/image/anim.png?raw=true"

# Download the image bytes from the link
resp = requests.get(url, stream=True).raw

# Read bytes and turn them into a NumPy array of type uint8
arr = np.asarray(bytearray(resp.read()), dtype="uint8")

# Decode the byte array into a color image (BGR format)
img = cv2.imdecode(arr, cv2.IMREAD_COLOR)
```

## Step 3: Check basic image info and display it
We print simple facts about the image, then convert BGR to RGB so it looks correct in Matplotlib.

```python
# Print the image size as (height, width, channels)
print("Shape:", img.shape)

# Print the data type used for pixel values
print("Data type:", img.dtype)

# Print the smallest pixel value in the image
print("Min pixel value:", img.min())

# Print the biggest pixel value in the image
print("Max pixel value:", img.max())

# Convert BGR image to RGB so colors look correct in Matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show the RGB image
plt.imshow(img_rgb)

# Add a title above the image
plt.title("Original Image (RGB)")

# Hide x-axis and y-axis numbers
plt.axis("off")

# Render the figure on screen
plt.show()
```

## Step 4: Split into Red, Green, and Blue channels
Each channel is a grayscale map that shows how strong that color is in each pixel.

```python
# Split the color image into blue, green, and red channels
b, g, r = cv2.split(img)

# Create a figure window with a wide size
plt.figure(figsize=(10, 5))

# Select subplot position 1 (first panel)
plt.subplot(1, 3, 1)

# Display the red channel as a grayscale image
plt.imshow(r, cmap='gray')

# Add title for the red channel panel
plt.title("Red channel strength")

# Hide axis for the red channel panel
plt.axis("off")

# Select subplot position 2 (middle panel)
plt.subplot(1, 3, 2)

# Display the green channel as a grayscale image
plt.imshow(g, cmap='gray')

# Add title for the green channel panel
plt.title("Green channel strength")

# Hide axis for the green channel panel
plt.axis("off")

# Select subplot position 3 (last panel)
plt.subplot(1, 3, 3)

# Display the blue channel as a grayscale image
plt.imshow(b, cmap='gray')

# Add title for the blue channel panel
plt.title("Blue channel strength")

# Hide axis for the blue channel panel
plt.axis("off")

# Show all three channel panels
plt.show()
```

## Step 5: Build pure red/green/blue images
We keep one channel and set the other two channels to zero.

```python
# Create a matrix of zeros with the same size as one channel
z = np.zeros_like(r)

# Build a BGR image that keeps only red values
red_bgr = cv2.merge([z, z, r])

# Build a BGR image that keeps only green values
green_bgr = cv2.merge([z, g, z])

# Build a BGR image that keeps only blue values
blue_bgr = cv2.merge([b, z, z])

# Convert red-only image from BGR to RGB for display
red = cv2.cvtColor(red_bgr, cv2.COLOR_BGR2RGB)

# Convert green-only image from BGR to RGB for display
green = cv2.cvtColor(green_bgr, cv2.COLOR_BGR2RGB)

# Convert blue-only image from BGR to RGB for display
blue = cv2.cvtColor(blue_bgr, cv2.COLOR_BGR2RGB)

# Create a figure window for the three color-only images
plt.figure(figsize=(10, 5))

# Select subplot position 1
plt.subplot(1, 3, 1)

# Show red-only image
plt.imshow(red)

# Add title for red-only image
plt.title("Only Red")

# Hide axis for red-only image
plt.axis("off")

# Select subplot position 2
plt.subplot(1, 3, 2)

# Show green-only image
plt.imshow(green)

# Add title for green-only image
plt.title("Only Green")

# Hide axis for green-only image
plt.axis("off")

# Select subplot position 3
plt.subplot(1, 3, 3)

# Show blue-only image
plt.imshow(blue)

# Add title for blue-only image
plt.title("Only Blue")

# Hide axis for blue-only image
plt.axis("off")

# Show all three color-only images
plt.show()
```

## Step 6: Make grayscale in two ways and compare
Method A: Custom formula (more weight to red here).
Method B: OpenCV built-in grayscale conversion.

```python
# Make a custom grayscale image using weighted red, green, and blue
custom = (0.6 * r) + (0.3 * g) + (0.1 * b)

# Convert custom image values to uint8 pixel format
custom = custom.astype(np.uint8)

# Make grayscale image using OpenCV built-in method
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Create a figure window for side-by-side comparison
plt.figure(figsize=(10, 5))

# Select subplot position 1
plt.subplot(1, 2, 1)

# Show custom grayscale image
plt.imshow(custom, cmap='gray')

# Add title for custom grayscale image
plt.title("Custom Grayscale")

# Hide axis for custom grayscale image
plt.axis("off")

# Select subplot position 2
plt.subplot(1, 2, 2)

# Show OpenCV grayscale image
plt.imshow(gray, cmap='gray')

# Add title for OpenCV grayscale image
plt.title("OpenCV Grayscale")

# Hide axis for OpenCV grayscale image
plt.axis("off")

# Show both grayscale images
plt.show()
```
