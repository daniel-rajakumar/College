```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import requests

url = "https://raymondmcguire.github.io/GPU-Based-Image-Processing-Tools/tools_webgl/FilterViewer/image/anim.png?raw=true"
resp = requests.get(url, stream=True).raw
arr = np.asarray(bytearray(resp.read()), dtype="uint8")
img = cv2.imdecode(arr, cv2.IMREAD_COLOR)

print(img.shape)
print(img.dtype)
print(img.min())
print(img.max())

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
b, g, r = cv2.split(img)

plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(r, cmap='gray')
plt.subplot(1, 3, 2)
plt.imshow(g, cmap='gray')
plt.subplot(1, 3, 3)
plt.imshow(b, cmap='gray')
plt.show()

z = np.zeros_like(r)
red = cv2.merge([r, z, z])
green = cv2.merge([z, g, z])
blue = cv2.merge([z, z, b])

plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(red)
plt.subplot(1, 3, 2)
plt.imshow(green)
plt.subplot(1, 3, 3)
plt.imshow(blue)
plt.show()

custom = (0.6 * r) + (0.3 * g) + (0.1 * b)
custom = custom.astype(np.uint8)


gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(custom, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(gray, cmap='gray')
plt.show()
```
